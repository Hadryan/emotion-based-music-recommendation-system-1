```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2

from deepface import DeepFace
```

```python
image = cv2.imread('happy1.jpeg')
plt.imshow(image),plt.axis("off")
# by default cv2 takes BGR clor scheme
```

```python
# we can convert that by
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image),plt.axis("off")
```

```python
image2 = image
predict = DeepFace.analyze(image2)
```

```python
import cv2
from deepface import DeepFace

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Unable to access cam..")

while True:
    ret,frame = cap.read()
    result = DeepFace.analyze(frame,actions=['emotion'])
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray,1.1,4)
    for(x, y, w, h) in faces:
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
    font  = cv2.FONT_HERSHEY_SIMPLEX
    print(result['dominant_emotion'])
    cv2.putText(frame, result['dominant_emotion'], (200,170), font, 0.5, (0,0,255), 2,cv2.LINE_8)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
